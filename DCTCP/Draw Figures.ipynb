```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import subprocess
```

```python
#Draw figure for 1.3.1
data1=pd.read_csv("Hw4/queue-<dctcp>-<20>-<0.062500>-<200>.txt",sep=' ',header=None)
data2=pd.read_csv("Hw4/queue-<tcp>-<20>-<0.062500>-<200>.txt",sep=' ',header=None)

data1.columns=['Time','Length']
data2.columns=['Time','Length']

x1=data1['Time']
y1=data1['Length']
x2=data2['Time']
y2=data2['Length']


plt.figure(figsize=(10,10))
plt.plot(x1,y1,label="DCTCP")
plt.plot(x2,y2,label='TCP')
plt.title('Queue Length 1',fontsize=10)
plt.xlabel('Time',fontsize=30)
plt.ylabel('Queue Length(packets)',fontsize=10)
plt.legend()
```

```python
# Draw figure for 1.3.2
data1=pd.read_csv("Hw4/queue-<dctcp>-<20>-<0.062500>-<200>.txt",sep=' ',header=None)
data2=pd.read_csv("Hw4/queue-<tcp>-<20>-<0.062500>-<200>.txt",sep=' ',header=None)
def ecdf(data):
    x = np.sort(data)
    n = x.size
    y = np.arange(1, n+1) / n
    return(x,y)
dataset1=data1[1]
x1,y1=ecdf(dataset1)
dataset2=data2[1]
x2,y2=ecdf(dataset2)
plt.xlabel('Queue Length(packets)')
plt.ylabel('Cumulative Fraction')

plt.plot(x1,y1,label="DCTCP")
plt.plot(x2,y2,label="TCP")
plt.legend()
```
